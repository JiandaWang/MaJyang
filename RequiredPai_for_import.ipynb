```python
#--------------------------------------------------------------------------------------------
def DisplayPool(f_pool):  
    # loop over all PAIs in the pool
    for current_PAI in f_pool:
        # print the PAI
        print(g_PAI_Disp_c[current_PAI], end='')
    
    # return
    return

#--------------------------------------------------------------------------------------------
def Dist2PAIs(f_PAI_1, f_PAI_2): 
    # return the distance if the distance <= 3; otherwise return invalid value
    if g_PAI_Dict_c[f_PAI_1]//10 == g_PAI_Dict_c[f_PAI_2]//10:
        dist = abs(g_PAI_Dict_c[f_PAI_1] - g_PAI_Dict_c[f_PAI_2])
        if dist <= 3: return dist
        else: return g_invalid_PAI_Dist_c
    else: return g_invalid_PAI_Dist_c

#--------------------------------------------------------------------------------------------
def Dist2IDs(f_ID_1, f_ID_2):
    # if any input is invalid, return invalid output
    if not f_ID_1 in g_PAI_Reverse_Dict_c \
    or not f_ID_2 in g_PAI_Reverse_Dict_c:
        return g_invalid_PAI_Dist_c
    else: return Dist2PAIs(g_PAI_Reverse_Dict_c[f_ID_1], g_PAI_Reverse_Dict_c[f_ID_2])

#--------------------------------------------------------------------------------------------
def Seq(f_hand):
    # convert a PAI_list into a (coded) sequence for further process
    # basically only save the number of same PAI and distance between different PAIs
    # if the difference of the number is >2, than we define as infinite; same for distance between different sorts of PAIs
    
    # initialze return value(s)
    r_seq = []
    r_comp_Hand = []
    
    # initialize current_PAI (assuming always not empty)
    current_PAI = f_hand[0]
    PAI_Counter = 1
    
    # loop over all rest PAIs in input list
    for i in range(1, len(f_hand)):
        # iterator equals current PAI, increase the counter
        if current_PAI == f_hand[i]: 
            PAI_Counter += 1
        # otherwise push current PAI and distance into output sequence, replace current PAI with iterator
        else:
            r_seq.append(PAI_Counter)
            r_seq.append(Dist2PAIs(current_PAI, f_hand[i]))
            r_comp_Hand.append(current_PAI)
            current_PAI = f_hand[i]
            PAI_Counter = 1
            
    # push the last PAI into output sequence
    r_seq.append(PAI_Counter)
    r_comp_Hand.append(current_PAI)
    
    # return
    return r_seq, r_comp_Hand

#--------------------------------------------------------------------------------------------
def CalNumOfNecessaryPAIs(f_hand, fr_required_PAIs):
    # calculate number of necessary PAIs / 向听数 / シャンテン数
    
    # initialze return value(s)
    r_num = g_invalid_Num_Of_Necessary_PAIs_c

    # convert input hand into sequence
    l_hand_Seq, l_comp_Hand = Seq(f_hand)
        
    # calculate with normal approach (fully concealed / 门清 / メンゼンチン not required)
    r_num = CalNum4NormApproach(l_hand_Seq, l_comp_Hand, len(f_hand), r_num, fr_required_PAIs)
    
    # when the hand is fully concealed, check further special cases
    if len(f_hand) == g_num_PAIs_Fully_Concealed_Hand_c:
        # calculate for 7 pairs / 七对子 / チートイツ
        r_num = CalNum4SevenPairs(l_hand_Seq, l_comp_Hand, r_num, fr_required_PAIs)
    
        # calculate for thirteen orphans / 十三幺 / 国士無双
        r_num = CalNum4ThirteenOrphans(f_hand, r_num, fr_required_PAIs)
    
    # return
    return r_num

#--------------------------------------------------------------------------------------------
def CalNum4NormApproach(f_hand_Seq, f_comp_Hand, f_hand_Length, f_current_Num, fr_required_PAIs):
    # calculate the number of necessary PAIs for normal approach
    # initialize return value(s)
    r_num = f_current_Num

    #--------------------------------------------------------------------------------------------
    # calculate with pairs / 雀頭
    # initialize local variables
    l_hand_Seq = f_hand_Seq.copy()
    
    # loop over all possible pairs
    l_take_Pair_Possible, l_next_Start_Index = TryTakePair(l_hand_Seq, 0)
    
    while (False != l_take_Pair_Possible):        
        # try to take 3ers / 順子 / 刻子
        # current hand sequence, start with index 0, no possible list yet, max taken 3ers number 0, no 3ers already taken
        l_possible_Hand_Seqs_After_Taken_3ers = []
        l_max_3ers_Taken = TryTake3ers(l_hand_Seq, f_comp_Hand, \
                                       0, 0, 0, \
                                       [], \
                                       l_possible_Hand_Seqs_After_Taken_3ers)
        
        # for each possible hand sequence in the list where all possible 3ers have been taken, try to take 2ers
        l_max_2ers_Taken = 0
        l_2ers_Taken = set()
        l_single_Remaining_PAIs = set()
        
        for l_hand_Seq_After_Taken_3ers in l_possible_Hand_Seqs_After_Taken_3ers:
            l_max_2ers_Taken = TryTake2ers(l_hand_Seq_After_Taken_3ers[0], f_comp_Hand, \
                                           0, l_max_2ers_Taken, 0, \
                                           l_2ers_Taken, l_single_Remaining_PAIs, \
                                           set(), set())
            
            r_num = UpdateNum4NormalApproach(r_num, f_hand_Length, True, l_max_3ers_Taken, l_max_2ers_Taken,
                                             l_hand_Seq_After_Taken_3ers[1], l_2ers_Taken, l_single_Remaining_PAIs,
                                             fr_required_PAIs)           
               
        # reset the hand and try to get the next pair
        l_hand_Seq = f_hand_Seq.copy()
        l_take_Pair_Possible, l_next_Start_Index = TryTakePair(l_hand_Seq, l_next_Start_Index)
        
    #--------------------------------------------------------------------------------------------
    # calculate without pairs / 雀頭
    # reset the hand and try to get the next pair
    l_hand_Seq = f_hand_Seq.copy()
    # try to take 3ers / 順子 / 刻子
    # current hand sequence, start with index 0, no possible list yet, max taken 3ers number 0, no 3ers already taken
    l_possible_Hand_Seqs_After_Taken_3ers = []
    l_max_3ers_Taken = TryTake3ers(l_hand_Seq, f_comp_Hand, \
                                   0, 0, 0, \
                                   [], \
                                   l_possible_Hand_Seqs_After_Taken_3ers)

    # for each possible hand sequence in the list where all possible 3ers have been taken, try to take 2ers
    l_max_2ers_Taken = 0
    l_2ers_Taken = set()
    l_single_Remaining_PAIs = set()
    
    for l_hand_Seq_After_Taken_3ers in l_possible_Hand_Seqs_After_Taken_3ers:
        l_max_2ers_Taken = TryTake2ers(l_hand_Seq_After_Taken_3ers[0], f_comp_Hand, \
                                       0, l_max_2ers_Taken, 0, \
                                       l_2ers_Taken, l_single_Remaining_PAIs, \
                                       set(), set())

        r_num = UpdateNum4NormalApproach(r_num, f_hand_Length, False, l_max_3ers_Taken, l_max_2ers_Taken,
                                         l_hand_Seq_After_Taken_3ers[1], l_2ers_Taken, l_single_Remaining_PAIs,
                                         fr_required_PAIs)    
    # return
    # return also possible number of pairs for calculation for 7 pairs / 七对子 / チートイツ
    return r_num

#--------------------------------------------------------------------------------------------
def TryTakePair(fr_hand_Seq, f_start_Index):
    # initialize return value(s)
    r_take_Pair_Possible = False
    r_next_Start_Index = f_start_Index

    # search for possible pair
    while (r_next_Start_Index<len(fr_hand_Seq) and fr_hand_Seq[r_next_Start_Index]<2):
        r_next_Start_Index += 2 # due to the element inbetween is distance
    
    # in case valid pair has been found
    if r_next_Start_Index<len(fr_hand_Seq):
        r_take_Pair_Possible = True # taking pair is possible
        fr_hand_Seq[r_next_Start_Index] -= 2 # take the pair from the sequence
        r_next_Start_Index += 2 # set the start index for next loop
    
    # return
    return r_take_Pair_Possible, r_next_Start_Index

#--------------------------------------------------------------------------------------------
def TryTake3ers(f_hand_Seq, f_comp_Hand,
                f_start_Index, f_max_3ers_Taken, f_num_3ers_Taken, \
                f_current_3ers_Taken, \
                fr_current_Possible_Hand_Seqs_And_3ers_Taken):
    # use DFS to recursively take 3ers from the hand sequence
    # for each call (-layer) only take one 3er -> after taken one "directly" return
    #    -> at same layer no further investigation is necessary, 
    #       because by skipping this 3er the final number of taken 3ers is definitely not optimal 
    # speiciality for abc form:
    # 1: next call starts with same index 
    #    -> still other 3ers with this index possible (aaaabc will be covered in this case)
    # 2: if with current start index only abc could be found 
    #    -> start with "b" and "c" should be checked in the same layer(cover abcde, abbbc and so on)
    #    -> for aaa form not necessary because these would be covered in next call (-layer) (aaabcd and so on)
    
    # initialize local variable(s)
    l_hand_Seq = f_hand_Seq.copy()
    l_current_3ers_Taken = f_current_3ers_Taken.copy()
    l_max_3ers_Taken = f_max_3ers_Taken
    l_index_To_Check = f_start_Index
    l_end_Index = len(f_hand_Seq)-1
       
    # loop over all remaining PAIs
    while (l_index_To_Check <= l_end_Index):
        # check aaa form
        if l_hand_Seq[l_index_To_Check] >= 3:
            # take the aaa form 3ers from the hand sequence
            l_hand_Seq[l_index_To_Check] -= 3
            l_current_3ers_Taken.append(['aaa', f_comp_Hand[l_index_To_Check//2]])
            # update the end index -> further check on this layer not necessary
            l_end_Index = l_index_To_Check
            # try to take further 3ers
            l_max_3ers_Taken = TryTake3ers(l_hand_Seq, f_comp_Hand, \
                                           l_index_To_Check+2, l_max_3ers_Taken, f_num_3ers_Taken+1, \
                                           l_current_3ers_Taken, \
                                           fr_current_Possible_Hand_Seqs_And_3ers_Taken)
            # reset hand sequence
            l_hand_Seq = f_hand_Seq.copy()
            l_current_3ers_Taken = f_current_3ers_Taken.copy()
            
        # check abc form:
        if l_index_To_Check+4 < len(f_hand_Seq) \
        and l_hand_Seq[l_index_To_Check] > 0 \
        and l_hand_Seq[l_index_To_Check+1] == 1 \
        and l_hand_Seq[l_index_To_Check+2] > 0 \
        and l_hand_Seq[l_index_To_Check+3] == 1 \
        and l_hand_Seq[l_index_To_Check+4] > 0:
            # take the aaa form 3ers from the hand sequence
            l_hand_Seq[l_index_To_Check] -= 1
            l_hand_Seq[l_index_To_Check+2] -= 1
            l_hand_Seq[l_index_To_Check+4] -= 1
            l_current_3ers_Taken.append(['abc', f_comp_Hand[l_index_To_Check//2]])
            # update the end index -> index of "b" and "c" should be check on this layer to cover abcde case
            # if already aaa form is found for this index, then it's not necessary
            l_end_Index = min(l_index_To_Check+4, l_end_Index)
            # try to take further 3ers -> start with same index
            l_max_3ers_Taken = TryTake3ers(l_hand_Seq, f_comp_Hand, \
                                           l_index_To_Check, l_max_3ers_Taken, f_num_3ers_Taken+1, \
                                           l_current_3ers_Taken, \
                                           fr_current_Possible_Hand_Seqs_And_3ers_Taken)
            # reset hand sequence
            l_hand_Seq = f_hand_Seq.copy()
            l_current_3ers_Taken = f_current_3ers_Taken.copy()
            
        # increase index to check
        l_index_To_Check += 2 # values inbetween are distances
       
    # return -> ensured with the l_max_3ers_Taken
    return UpdateHandSeqTaken3ers(f_hand_Seq, l_max_3ers_Taken, f_num_3ers_Taken, \
                                  l_current_3ers_Taken, \
                                  fr_current_Possible_Hand_Seqs_And_3ers_Taken)

#--------------------------------------------------------------------------------------------
def UpdateHandSeqTaken3ers(f_hand_Seq, f_max_3ers_Taken, f_num_3ers_Taken, \
                           f_current_3ers_Taken, \
                           fr_current_Possible_Hand_Seqs_And_3ers_Taken):
    # replace the output with current hand sequence if more 3ers have been taken at this layer
    # if the numbers are the same, add the current hand sequence to the output (eliminate duplicated elements)
    # otherwise do no update
    if f_num_3ers_Taken > f_max_3ers_Taken:
        fr_current_Possible_Hand_Seqs_And_3ers_Taken.clear()
        fr_current_Possible_Hand_Seqs_And_3ers_Taken.append([f_hand_Seq, f_current_3ers_Taken])
        return f_num_3ers_Taken
    elif f_num_3ers_Taken == f_max_3ers_Taken \
    and not [f_hand_Seq, f_current_3ers_Taken] in fr_current_Possible_Hand_Seqs_And_3ers_Taken:
        fr_current_Possible_Hand_Seqs_And_3ers_Taken.append([f_hand_Seq, f_current_3ers_Taken]),
        return f_max_3ers_Taken
    else: return f_max_3ers_Taken

#--------------------------------------------------------------------------------------------
def TryTake2ers(f_hand_Seq, f_comp_Hand, f_start_Index, f_max_2ers_Taken, f_num_2ers_Taken, \
                fr_2ers_Taken, fr_single_Remaining_PAIs, \
                f_local_2ers_Taken, f_local_Single_Remaining_PAIs):
       
    # initialize local variable(s)
    l_hand_Seq = f_hand_Seq.copy()
    l_index_To_Check = f_start_Index # note: index of hand sequence / 2 = index of compressed hand
    l_end_Index = len(l_hand_Seq)
    l_max_2ers_Taken = f_max_2ers_Taken
    l_local_2ers_Taken = f_local_2ers_Taken.copy()
    l_local_Single_Remaining_PAIs = f_local_Single_Remaining_PAIs.copy()
    
    # loop over input hand sequence
    while (l_index_To_Check < l_end_Index):
        # check aa form:
        if l_hand_Seq[l_index_To_Check] > 1:
            # remove aa from the hand sequence
            l_hand_Seq[l_index_To_Check] -= 2
            # update the taken 2ers set
            l_local_2ers_Taken.add(('aa', f_comp_Hand[l_index_To_Check // 2]))
            # next search
            l_max_2ers_Taken = TryTake2ers(l_hand_Seq, f_comp_Hand, l_index_To_Check, l_max_2ers_Taken, f_num_2ers_Taken+1, \
                                           fr_2ers_Taken, fr_single_Remaining_PAIs, \
                                           l_local_2ers_Taken, l_local_Single_Remaining_PAIs)
            # stop the loop at this layer 
            l_end_Index = l_index_To_Check
            # reset hand sequence
            l_hand_Seq = f_hand_Seq.copy()
            # reset the taken 2ers set
            l_local_2ers_Taken = f_local_2ers_Taken.copy()
            
        # check ab form
        if l_index_To_Check+2 < len(l_hand_Seq) \
        and l_hand_Seq[l_index_To_Check] > 0 \
        and l_hand_Seq[l_index_To_Check+1] == 1 \
        and l_hand_Seq[l_index_To_Check+2] > 0:
            # remove av from the hand sequence
            l_hand_Seq[l_index_To_Check] -= 1
            l_hand_Seq[l_index_To_Check+2] -= 1
            # update the taken 2ers set
            l_local_2ers_Taken.add(('ab', f_comp_Hand[l_index_To_Check // 2]))
            # next search
            l_max_2ers_Taken = TryTake2ers(l_hand_Seq, f_comp_Hand, l_index_To_Check, l_max_2ers_Taken, f_num_2ers_Taken+1, \
                                           fr_2ers_Taken, fr_single_Remaining_PAIs, \
                                           l_local_2ers_Taken, l_local_Single_Remaining_PAIs)
            # update end index 
            l_end_Index = min(l_end_Index, l_index_To_Check+4)
            # reset hand sequence
            l_hand_Seq = f_hand_Seq.copy()
            # reset the taken 2ers set
            l_local_2ers_Taken = f_local_2ers_Taken.copy()
            
        # checj a_c form 1
        if l_index_To_Check+2 < len(l_hand_Seq) \
        and l_hand_Seq[l_index_To_Check] > 0 \
        and l_hand_Seq[l_index_To_Check+1] == 2 \
        and l_hand_Seq[l_index_To_Check+2] > 0:
            # remove ac from the hand sequence
            l_hand_Seq[l_index_To_Check] -= 1
            l_hand_Seq[l_index_To_Check+2] -= 1
            # update the taken 2ers set
            l_local_2ers_Taken.add(('ac', f_comp_Hand[l_index_To_Check // 2]))
            # next search
            l_max_2ers_Taken = TryTake2ers(l_hand_Seq, f_comp_Hand, l_index_To_Check, l_max_2ers_Taken, f_num_2ers_Taken+1, \
                                           fr_2ers_Taken, fr_single_Remaining_PAIs, \
                                           l_local_2ers_Taken, l_local_Single_Remaining_PAIs)
            # update end index 
            l_end_Index = min(l_end_Index, l_index_To_Check+4)
            # reset hand sequence
            l_hand_Seq = f_hand_Seq.copy()        
            # reset the taken 2ers
            l_local_2ers_Taken = f_local_2ers_Taken.copy()
            
        # check a_c form 2
        if l_index_To_Check+4 < len(l_hand_Seq) \
        and l_hand_Seq[l_index_To_Check] > 0 \
        and l_hand_Seq[l_index_To_Check+1] == 1 \
        and l_hand_Seq[l_index_To_Check+2] == 0 \
        and l_hand_Seq[l_index_To_Check+3] == 1 \
        and l_hand_Seq[l_index_To_Check+4] > 0:
            # remove ac from the hand sequence
            l_hand_Seq[l_index_To_Check] -= 1
            l_hand_Seq[l_index_To_Check+4] -= 1
            # update the taken 2ers set
            l_local_2ers_Taken.add(('ac', f_comp_Hand[l_index_To_Check // 2]))
            # next search
            l_max_2ers_Taken = TryTake2ers(l_hand_Seq, f_comp_Hand, l_index_To_Check, l_max_2ers_Taken, f_num_2ers_Taken+1, \
                                           fr_2ers_Taken, fr_single_Remaining_PAIs, \
                                           l_local_2ers_Taken, l_local_Single_Remaining_PAIs)
            # update end index 
            l_end_Index = min(l_end_Index, l_index_To_Check+6)
            # reset hand sequence
            l_hand_Seq = f_hand_Seq.copy()
            # reset the taken 2ers set
            l_local_2ers_Taken = f_local_2ers_Taken.copy()

        # nothing found / search on this layer ended
        # update the required PAI set
        if (l_hand_Seq[l_index_To_Check] > 0):
            l_local_Single_Remaining_PAIs.add(f_comp_Hand[l_index_To_Check // 2])
        # increase the index to check next element, 
        l_index_To_Check += 2;
       
    # return -> ensured with the l_max_2ers_Taken
    return UpdateRequiredPAIsTaken2ers(l_max_2ers_Taken, f_num_2ers_Taken, \
                                       fr_2ers_Taken, fr_single_Remaining_PAIs, \
                                       l_local_2ers_Taken, l_local_Single_Remaining_PAIs)

#--------------------------------------------------------------------------------------------
def UpdateRequiredPAIsTaken2ers(f_max_2ers_Taken, f_num_2ers_Taken, \
                                fr_2ers_Taken, fr_single_Remaining_PAIs, \
                                f_local_2ers_Taken, f_local_Single_Remaining_PAIs):   
    # replace the required PAI set when more 2ers have been taken
    # if the numbers are the same, merge the local required PAIs set to output
    # else do nothing
    if f_num_2ers_Taken > f_max_2ers_Taken:
        fr_2ers_Taken.clear(); fr_2ers_Taken |= f_local_2ers_Taken
        fr_single_Remaining_PAIs.clear(); fr_single_Remaining_PAIs |= f_local_Single_Remaining_PAIs
        return f_num_2ers_Taken
    elif f_num_2ers_Taken == f_max_2ers_Taken:
        fr_2ers_Taken |= f_local_2ers_Taken
        fr_single_Remaining_PAIs |= f_local_Single_Remaining_PAIs
        return f_num_2ers_Taken
    else: return f_max_2ers_Taken    

#--------------------------------------------------------------------------------------------
def GetRequiredPAIsSplitting3ers(f_form, f_PAI_1, \
                                 f_available_Single_PAIs = {'1m', '2m', '3m', '4m', '5m', '6m', '7m', '8m', '9m', \
                                                            '1p', '2p', '3p', '4p', '5p', '6p', '7p', '8p', '9p', \
                                                            '1s', '2s', '3s', '4s', '5s', '6s', '7s', '8s', '9s', \
                                                            'E', 'S', 'W', 'N', 'R', 'G', 'Wh'}):
    # when 2ers are not enough while calculating the necessary PAIs, splitting 3ers + alpha into two 2ers would also help
    # alpha should be inlcuded in the input set of available single PAIs
    # return required PAIs in this case
    
    # initialize return value(s)
    r_set_Of_Required_PAIs = set()
    l_ID_1 = g_PAI_Dict_c[f_PAI_1]
    
    # 'aaa' form: check y, z, b, c
    if f_form == 'aaa':
        # check yaaa -> split into ac and aa
        if l_ID_1-2 in g_PAI_Reverse_Dict_c \
        and g_PAI_Reverse_Dict_c[l_ID_1-2] in f_available_Single_PAIs \
        and 2 == Dist2IDs(l_ID_1, l_ID_1-2):
            r_set_Of_Required_PAIs |= GetRequiredPAIs('ac', g_PAI_Reverse_Dict_c[l_ID_1-2])
            r_set_Of_Required_PAIs |= GetRequiredPAIs('aa', f_PAI_1)
    
        # check zaaa -> split into ab and aa
        if l_ID_1-1 in g_PAI_Reverse_Dict_c \
        and g_PAI_Reverse_Dict_c[l_ID_1-1] in f_available_Single_PAIs \
        and 1 == Dist2IDs(l_ID_1, l_ID_1-1):
            r_set_Of_Required_PAIs |= GetRequiredPAIs('ab', g_PAI_Reverse_Dict_c[l_ID_1-1])
            r_set_Of_Required_PAIs |= GetRequiredPAIs('aa', f_PAI_1)
        
        # check aaab -> split into ab and aa
        if l_ID_1+1 in g_PAI_Reverse_Dict_c \
        and g_PAI_Reverse_Dict_c[l_ID_1+1] in f_available_Single_PAIs \
        and 1 == Dist2IDs(l_ID_1, l_ID_1+1):
            r_set_Of_Required_PAIs |= GetRequiredPAIs('ab', f_PAI_1)
            r_set_Of_Required_PAIs |= GetRequiredPAIs('aa', f_PAI_1)
        
        # check aaac -> split into ac and aa
        if l_ID_1+2 in g_PAI_Reverse_Dict_c \
        and g_PAI_Reverse_Dict_c[l_ID_1+2] in f_available_Single_PAIs \
        and 2 == Dist2IDs(l_ID_1, l_ID_1+2):
            r_set_Of_Required_PAIs |= GetRequiredPAIs('ac', f_PAI_1)
            r_set_Of_Required_PAIs |= GetRequiredPAIs('aa', f_PAI_1)
    
    elif f_form == 'abc':
        # check yabc -> split into ac and ab
        if l_ID_1-2 in g_PAI_Reverse_Dict_c \
        and g_PAI_Reverse_Dict_c[l_ID_1-2] in f_available_Single_PAIs \
        and 2 == Dist2IDs(l_ID_1, l_ID_1-2):
            r_set_Of_Required_PAIs |= GetRequiredPAIs('ac', g_PAI_Reverse_Dict_c[g_PAI_Dict_c[f_PAI_1]-2])
            r_set_Of_Required_PAIs |= GetRequiredPAIs('ab', g_PAI_Reverse_Dict_c[g_PAI_Dict_c[f_PAI_1]+1])

        # check zabc -> split into ab and ab (or ac and ac, covered by ab and ab form)
        if l_ID_1-1 in g_PAI_Reverse_Dict_c \
        and g_PAI_Reverse_Dict_c[g_PAI_Dict_c[f_PAI_1]-1] in f_available_Single_PAIs \
        and 1 == Dist2IDs(l_ID_1, l_ID_1-1):
            r_set_Of_Required_PAIs |= GetRequiredPAIs('ab', g_PAI_Reverse_Dict_c[l_ID_1-1])
            r_set_Of_Required_PAIs |= GetRequiredPAIs('ab', g_PAI_Reverse_Dict_c[l_ID_1+1])
        
        # check aabc -> split into (aa) and ab or ab and ac
        # merge next two cases together after review
        if f_PAI_1 in f_available_Single_PAIs \
        or g_PAI_Reverse_Dict_c[l_ID_1+1] in f_available_Single_PAIs \
        or g_PAI_Reverse_Dict_c[l_ID_1+2] in f_available_Single_PAIs:
            r_set_Of_Required_PAIs |= GetRequiredPAIs('ab', g_PAI_Reverse_Dict_c[l_ID_1+1])
            r_set_Of_Required_PAIs |= GetRequiredPAIs('ab', f_PAI_1)
            r_set_Of_Required_PAIs |= GetRequiredPAIs('ac', f_PAI_1)

        # check abbc -> split into ab and ab or (aa) and ac
#         if g_PAI_Reverse_Dict_c[g_PAI_Dict_c[f_PAI_1]+1] in f_available_Single_PAIs:
#             r_set_Of_Required_PAIs |= GetRequiredPAIs('ab', f_PAI_1)
#             r_set_Of_Required_PAIs |= GetRequiredPAIs('ab', g_PAI_Reverse_Dict_c[g_PAI_Dict_c[f_PAI_1]+1])
#             r_set_Of_Required_PAIs |= GetRequiredPAIs('ac', f_PAI_1)
        
        # check abcc -> split into ab and (aa) or ac and ab
#         if g_PAI_Reverse_Dict_c[g_PAI_Dict_c[f_PAI_1]+2] in f_available_Single_PAIs:
#             r_set_Of_Required_PAIs |= GetRequiredPAIs('ab', f_PAI_1)
#             r_set_Of_Required_PAIs |= GetRequiredPAIs('ac', f_PAI_1)
#             r_set_Of_Required_PAIs |= GetRequiredPAIs('ab', g_PAI_Reverse_Dict_c[g_PAI_Dict_c[f_PAI_1]+1])

        # check abcd -> split into ab and ab (or ac and ac, covered by ab and ab form)
        if l_ID_1+3 in g_PAI_Reverse_Dict_c \
        and g_PAI_Reverse_Dict_c[l_ID_1+3] in f_available_Single_PAIs \
        and 1 == Dist2IDs(l_ID_1+2, l_ID_1+3):
            r_set_Of_Required_PAIs |= GetRequiredPAIs('ab', f_PAI_1)
            r_set_Of_Required_PAIs |= GetRequiredPAIs('ab', g_PAI_Reverse_Dict_c[l_ID_1+2])

        # check abce -> split into ab and ac
        if l_ID_1+4 in g_PAI_Reverse_Dict_c \
        and g_PAI_Reverse_Dict_c[g_PAI_Dict_c[f_PAI_1]+4] in f_available_Single_PAIs \
        and 2 == Dist2IDs(l_ID_1+2, l_ID_1+4):
            r_set_Of_Required_PAIs |= GetRequiredPAIs('ab', f_PAI_1)
            r_set_Of_Required_PAIs |= GetRequiredPAIs('ac', g_PAI_Reverse_Dict_c[l_ID_1+2])
        
    # return
    return r_set_Of_Required_PAIs

#--------------------------------------------------------------------------------------------
def UpdateNum4NormalApproach(f_num, f_hand_Length, f_pair_Available, f_max_3ers_Taken, f_max_2ers_Taken,
                             f_3ers_Taken, f_2ers_Taken, f_single_Remaining_PAIs,
                             fr_required_PAIs):
    # initialize return value(s)
    r_num = f_num
    
    # initialize local variable(s) 
    l_2ers_Remaining = f_max_2ers_Taken
    l_hand_Length = f_hand_Length
    
    # calculate with pair
    if f_pair_Available:
        # calculate the remaining length
        l_hand_Length -= 2 + 3 * f_max_3ers_Taken # remaining 3n-1
        # calculate the nessary numbers of 2ers
        l_necessary_Num_2ers = l_hand_Length // 3 + 1
        # calculate the output number
        if f_max_2ers_Taken >= l_necessary_Num_2ers: l_num = l_necessary_Num_2ers - 1
        else: l_num = f_max_2ers_Taken \
                     +(l_necessary_Num_2ers - f_max_2ers_Taken - 1)*2 \
                     +1
        # update the outputs
        if l_num <= r_num:
            # update number and reset the output if necessary
            if l_num < r_num: r_num = l_num; fr_required_PAIs.clear()
            # always consider 2ers
            for l_2er in f_2ers_Taken: 
                fr_required_PAIs |= GetRequiredPAIs(l_2er[0], l_2er[1])
            # in case 2ers are not enough, consider also single PAIs and splitting 3ers
            if f_max_2ers_Taken < l_necessary_Num_2ers:
                for l_single in f_single_Remaining_PAIs: fr_required_PAIs |= GetRequiredPAIs('a', l_single)
                for l_3er in f_3ers_Taken: fr_required_PAIs |= GetRequiredPAIsSplitting3ers(l_3er[0], l_3er[1], \
                                                                                            f_single_Remaining_PAIs)
        
    # calculate without pair:
    else:    
        # calculate the remaining length
        l_hand_Length -= 3 * f_max_3ers_Taken # remaining 3n+1
        # calculate the nessary numbers of 2ers
        l_necessary_Num_2ers = l_hand_Length // 3
        # calculate the output number
        if f_max_2ers_Taken >= l_necessary_Num_2ers: l_num = l_necessary_Num_2ers
        else: l_num = f_max_2ers_Taken \
                     +(l_necessary_Num_2ers - f_max_2ers_Taken)*2
        
        # update the output
        if l_num <= r_num:
            # update number and reset the output if necessary
            if l_num < r_num: r_num = l_num; fr_required_PAIs.clear()
            # in case 2ers still necessary, consider 2ers
            if l_necessary_Num_2ers > 0: 
                for l_2er in f_2ers_Taken: fr_required_PAIs |= GetRequiredPAIs(l_2er[0], l_2er[1])
            # consider always single PAIs for pair:
            fr_required_PAIs |= f_single_Remaining_PAIs
            # if 2ers are too much, consider as single PAIs also for pair
            if f_max_2ers_Taken > l_necessary_Num_2ers: 
                fr_required_PAIs |= GetRequiredPAIs(l_2er[0]+'x', l_2er[1])
            # in case 2ers are not enough, consider also single PAIs and splitting 3ers
            elif f_max_2ers_Taken < l_necessary_Num_2ers:
                for l_single in f_single_Remaining_PAIs: fr_required_PAIs |= GetRequiredPAIs('a', l_single)
                for l_3er in f_3ers_Taken: fr_required_PAIs |= GetRequiredPAIsSplitting3ers(l_3er[0], l_3er[1], \
                                                                                              f_single_Remaining_PAIs)
            
    # return the minimal value between input and calculated new value
    return r_num

#--------------------------------------------------------------------------------------------
def CalNum4SevenPairs(f_hand_Seq, f_comp_Hand, f_num, fr_required_PAIs):
    # initialize return variable(s)
    r_num = f_num
    
    # initialize local variable(s)
    l_index_To_Check = 0
    l_pair_Count = 0
    l_single_Remaining_PAIs = set()
    
    # loop over all hand PAIs
    while l_index_To_Check < len(f_hand_Seq):
        # increase count if pair found; otherwise add to single remaining PAIs set
        if f_hand_Seq[l_index_To_Check] > 1: l_pair_Count += 1
        else: l_single_Remaining_PAIs.add(f_comp_Hand[l_index_To_Check // 2])
        # increase index
        l_index_To_Check += 2
    
    # calculate the number
    l_num = 6 - l_pair_Count
    
    # update the output
    if l_num <= r_num:
        # reset the output if necessary
        if l_num < r_num: r_num = l_num; fr_required_PAIs.clear()
        # update required PAIs set
        fr_required_PAIs |= l_single_Remaining_PAIs
    
    # return
    return r_num

#--------------------------------------------------------------------------------------------
def CalNum4ThirteenOrphans(f_hand, f_num, fr_required_PAIs):
    # intialize return value(s)
    r_num = f_num
    
    # initialize local variable(s)
    l_exist_Orphans = set()
    l_orphan_Pair_Available = False
    
    # loop over all PAIs in hand
    for PAI in f_hand:
        # check for orphan:
        if PAI in g_orphan_Pool_c:
            # check if any pair is already found -> Thirteen Orphans allows only one pair
            if not PAI in l_exist_Orphans: l_exist_Orphans.add(PAI)
            else: l_orphan_Pair_Available = True
    
    # calculate the number of necessary PAIs
    if not l_orphan_Pair_Available: l_num = 13 - len(l_exist_Orphans)
    else: l_num = 12 - len(l_exist_Orphans) # one pair is allowed for thirteen orphans
    
    # update the output
    if l_num <= r_num:
        # reset the output if necessary
        if l_num < r_num: r_num = l_num; fr_required_PAIs.clear()
        # if no pair available, all orphans are required; otherwise only those missing ones
        if l_orphan_Pair_Available: fr_required_PAIs |= g_orphan_Pool_c - l_exist_Orphans
    
    # return the minimal value between input and calculated new value
    return r_num

#--------------------------------------------------------------------------------------------
def GetRequiredPAIs(f_form, f_PAI_1):
    # return required PAIs / 有效进张 / 有効牌
    # initialze return value(s)
    r_set_Of_Required_PAIs = set()
    
    # check form
    if f_form == 'a': # single PAI
        # get ID of PAI 1
        l_ID_1 = g_PAI_Dict_c[f_PAI_1]
        # check related PAIs
        r_set_Of_Required_PAIs.add(f_PAI_1)
        if 2 == Dist2IDs(l_ID_1, l_ID_1-2): r_set_Of_Required_PAIs.add(g_PAI_Reverse_Dict_c[l_ID_1-2])
        if 1 == Dist2IDs(l_ID_1, l_ID_1-1): r_set_Of_Required_PAIs.add(g_PAI_Reverse_Dict_c[l_ID_1-1])
        if 1 == Dist2IDs(l_ID_1, l_ID_1+1): r_set_Of_Required_PAIs.add(g_PAI_Reverse_Dict_c[l_ID_1+1])
        if 2 == Dist2IDs(l_ID_1, l_ID_1+2): r_set_Of_Required_PAIs.add(g_PAI_Reverse_Dict_c[l_ID_1+2])
        
    elif f_form == 'aa' or f_form == 'aax':
        r_set_Of_Required_PAIs.add(f_PAI_1)

    elif f_form == 'ab':
        # get ID of PAI 1
        l_ID_1 = g_PAI_Dict_c[f_PAI_1]
        if 1 == Dist2IDs(l_ID_1, l_ID_1-1): r_set_Of_Required_PAIs.add(g_PAI_Reverse_Dict_c[l_ID_1-1])
        if 2 == Dist2IDs(l_ID_1, l_ID_1+2): r_set_Of_Required_PAIs.add(g_PAI_Reverse_Dict_c[l_ID_1+2])        
            
    elif f_form == 'ac':
        # get ID of PAI 1
        l_ID_1 = g_PAI_Dict_c[f_PAI_1]
        if 1 == Dist2IDs(l_ID_1, l_ID_1+1): r_set_Of_Required_PAIs.add(g_PAI_Reverse_Dict_c[l_ID_1+1])
            
    elif f_form == 'abx':
        l_ID_1 = g_PAI_Dict_c[f_PAI_1]
        r_set_Of_Required_PAIs.add(f_PAI_1)
        r_set_Of_Required_PAIs.add(g_PAI_Reverse_Dict_c[l_ID_1+1])
        
    elif f_form == 'acx':
        l_ID_1 = g_PAI_Dict_c[f_PAI_1]
        r_set_Of_Required_PAIs.add(f_PAI_1)
        r_set_Of_Required_PAIs.add(g_PAI_Reverse_Dict_c[l_ID_1+2])        
        
    # return
    return r_set_Of_Required_PAIs

#--------------------------------------------------------------------------------------------
def GetRemainingNumOfRequiredPAIs(f_CurPAINum_cdict, f_required_PAIs):
    # initialize return value(s)
    r_num = 0
    
    # loop over all required PAIs and count the number
    for PAI in f_required_PAIs: r_num += f_CurPAINum_cdict[PAI]

    # return
    return r_num
```
